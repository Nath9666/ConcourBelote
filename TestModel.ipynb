```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from DetectAndReconizeDigits import detect_digits
from sklearn.preprocessing import MinMaxScaler

# Charger le modèle MNIST pré-entraîné
name_model = './models/mnist_model.keras'
svc_model = './models/svhn_model.keras'


# Charger et prétraiter l'image
image_path = './data/picture2.jpg'
img, digit_rois = detect_digits(image_path)

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (28, 28))
image = image / 255.0  # Normaliser les valeurs des pixels à [0, 1]


# Redimensionner l'image pour qu'elle corresponde à l'entrée attendue du modèle
image = np.reshape(image, (1, 28, 28, 1))

# Chargement du model
mnist_model = tf.keras.models.load_model(name_model)
digits = []

for (x, y, w, h, roi) in digit_rois:
    roi = cv2.resize(roi, (28, 28), interpolation=cv2.INTER_AREA)
    roi = roi.reshape(1, -1)
    roi = MinMaxScaler().fit_transform(roi)
    roi = roi.reshape(1, 28, 28, 1)  # Correction ici
    digit = mnist_model.predict(roi)
    print(f'Prediction: {digit}')
    digits.append((x, y, digit[0]))
    cv2.putText(img, str(np.argmax(digit)), (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 2)
    cv2.imshow("Digits Recognized", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Prédire le chiffre
predictions = mnist_model.predict(image)
predicted_digit = np.argmax(predictions)

print(f'Le chiffre reconnu est : {predicted_digit}')
```
